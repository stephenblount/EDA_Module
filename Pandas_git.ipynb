```python
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String
from sqlalchemy import inspect
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
engine = create_engine('sqlite:////home/stephen/Desktop/EDA/Pandas_test/mta_data_pd.db')
insp = inspect(engine)
print(insp.get_table_names())
```

```python
df = pd.read_sql('SELECT*FROM mta_data', engine)
```

```python
df['UNIQUE_ID'] = df['C/A'] + ' ' + df['UNIT'] + ' ' + df['SCP']
```

```python
df = df.sort_values(['UNIQUE_ID', 'DATE']).reset_index()
df.head()
```

```python
df['NET_ENTRIES'] = df['ENTRIES'] - df['ENTRIES'].shift(1)
df['NET_EXITS'] = df['EXITS'] - df['EXITS'].shift(1)
df.head(20)
```

```python
mask = ((df['NET_ENTRIES'] > 20000) | (df['NET_ENTRIES'] < 0) | (df['NET_EXITS'] > 20000) | (df['NET_EXITS'] < 0))
df = df[~mask]
```

```python
df1 = df.groupby('STATION')[['NET_ENTRIES', 'NET_EXITS']].sum().sort_values('NET_ENTRIES', ascending=False)
df1.head(20)
```

```python
df1['POPULATION'] = df1.NET_ENTRIES + df1.NET_EXITS
df1.head(5)
```
